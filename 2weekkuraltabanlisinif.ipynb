```python
from itertools import groupby

from pandas import read_csv
import numpy as np
import seaborn as sns
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

df = pd.read_csv('/Users/ry/PycharmProjects/pythonProject/pythonProgramlama/python_for_data_science/KuralTabanl覺S覺n覺fland覺rmaProje/persona.csv')

df.head()
df.nunique()
df["SOURCE"].value_counts()
df["PRICE"].value_counts()
df["COUNTRY"].value_counts()
df.groupby(["COUNTRY"]).agg({"PRICE": ["count","sum","min","max","mean"]})
df.groupby(["SOURCE"]).agg({"PRICE": ["count","sum","min","max","mean"]})
df.groupby(["SOURCE","COUNTRY"]).agg({"PRICE": ["count","sum","min","max","mean"]})

df.dtypes
grouped = df.groupby(["SOURCE","COUNTRY","AGE","SEX"]).agg({"PRICE": ["mean"]}).round(2)
grouped.dtypes

agg_df_temp = grouped.reset_index()
agg_df_temp.round(0)
agg_df = agg_df_temp.sort_values(by=('PRICE', 'mean'), ascending=False)

agg_df.head()
df_test= agg_df.reset_index().head()

agg_df.reset_index(level='INDEX', col_level=1)

val = np.squeeze(agg_df.values).T
agg_df_son = pd.DataFrame(val, [x[0] for x in agg_df.columns]).T
agg_df_son.head()
agg_df  = agg_df_son
agg_df.head()

bins = [0,18,23,30,40,agg_df["AGE"].max()]
titles = ['0_18','19_23','24_30','31_40','41_'+str(agg_df["AGE"].max())]
agg_df["AGE_CAT"] = pd.cut(agg_df["AGE"], bins, labels=titles)
agg_df.head()

#level_based_cust = [a.upper()+"_"+b.upper()+"_"c.upper()+"_"+ d for a,b,c,d] in zip(agg_df["COUNTRY"],agg_df["SOURCE"],agg_df["SEX"],agg_df["AGE_CAT"].astype(str))]
#customer_level_based = (agg_df["COUNTRY"].astype(str) + "_" + agg_df["SOURCE"].astype(str)+ "_" +agg_df["SEX"].astype(str)+ "_" +agg_df["AGE_CAT"].astype(str)).upper()

agg_df["customer_level_based"] = agg_df["COUNTRY"].str.cat(agg_df[["SOURCE", "SEX","AGE_CAT"]].astype(str), sep="_").str.upper()
agg_df.head()

new_agg_df = agg_df[["customer_level_based","PRICE"]]

agg_df = new_agg_df.groupby("customer_level_based")["PRICE"].mean()
agg_df = new_agg_df.groupby("customer_level_based").agg({"PRICE":"mean"})

agg_df = agg_df.reset_index()

agg_df["SEGMENT"] = pd.qcut(agg_df["PRICE"], 4, labels=["D","C","B","A"])
agg_df.groupby("SEGMENT").agg({"PRICE": ["mean","max","sum"]})

new_usr = "TUR_ANDROID_FEMALE_31_40"
agg_df[agg_df["customer_level_based"] == new_usr]
new_usr2 = "TUR_ANDROID_FEMALE_31_460"
agg_df[agg_df["customer_level_based"] == new_usr2]
```
